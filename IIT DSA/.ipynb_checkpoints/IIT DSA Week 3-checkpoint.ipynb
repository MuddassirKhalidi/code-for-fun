```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class doubly_linked_list:
    def __init__(self):
        self.head = None
        self.last = None
        
    def insert_data(self, data):
        node = Node(data)
        self.last.next = node
        node.prev = self.last
        self.last = node
        
        
    def delete_end(self):
        self.last = self.last.prev
        self.last.next = None
        
```

```python
class Hashing:
    def __init__(self,c1,c2,m):
        self.hashtable = []
        for i in range(m):
            self.hashtable.append(None)     
        self.c1 = c1
        self.c2 = c2
        self.m = m
        

    def store_data(self, data):
        initial_index = data % self.m
        
        for i in range(self.m):  # Quadratic probing iteration
            
            index = (initial_index + self.c1 * i + self.c2 * i * i) % self.m
            
            if self.hashtable[index] is None:  # If the slot is empty
                self.hashtable[index] = data  # Store the data
                return
            
        print("Hash table is full")  # If no empty slot is found after m attempts

    def display_hashtable(self):
        return self.hashtable  # Return the current state of the hashtable

    
```

```python
c1 = int(input())
c2 = int(input())
m = int(input())
data=eval(input())
A = Hashing(c1,c2,m)
for i in data:
    A.store_data(i)
print(A.display_hashtable())
```

```python
def DishPrepareOrder(order_list):
    ids = {x:0 for x in range(1001, 1010)}
    for x in order_list:
        ids[x] += 1
        
    orders = []
    for ID, count in ids.items():
        orders.append((ID, count))
    
    orders.sort(key=lambda x: (-x[1], x[0]))
    prepare_order = [ID for ID, count in orders if count > 0]
    return prepare_order
```

```python
def EvaluateExpression(exp:str):
    operators = ['+', '-', '*', '/', '**']
    elements = exp.split()
    stack = []
    for x in elements:
        if x not in operators:
            stack.append(x)
        else:
            e1, e2 = stack.pop(), stack.pop()
            if x == '+':
                result = e2 + e1
                
            elif x == '-':
                result = e2 - e1
                
            elif x == '*':
                result = e2 * e1
            
            elif x == '/':
                if e1 == 0:
                    print('Zero Division not allowed')
                    return
                else:
                    result = e2 / e1
            
            elif x == '**':
                result = e2 ** e1
                
            stack.append(result)
            
            #Could also use 'eval(f'{e2}{x}{e1}') but this results in security concerns as any user input could be executed as code'
            
    return stack[0]
            
            
```

```python
print(float(EvaluateExpression(input())))
```
