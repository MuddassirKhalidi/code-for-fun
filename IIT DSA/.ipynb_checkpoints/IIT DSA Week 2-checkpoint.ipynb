```python
def binarySearchIndexandComparisons(L: list[int], k:int):
    left, right = 0, len(L) - 1
    count = 0
    while left <= right:
        count += 1
        mid = (left + right) // 2
        
        if L[mid] == k:
            return (True, count)
        
        elif L[mid] < k:
            left = mid + 1
        else:
            right = mid - 1
        
    return (False, count)     
```

```python
l = [2, 6, 8, 11, 17, 23, 33, 44, 46, 50, 65]
k = 11

binarySearchIndexandComparisons(l, k)
```

```python
def sortInRange(L, r):
    num_map = [0] * r
    for x in L:
        num_map[x] += 1
        
    index = 0  
    
    for element in range(r):  
        count = num_map[element]  
        for _ in range(count):    
            L[index] = element
            index += 1
```

```python
def combinationSort(L):
    lst1 = mergesort_by_character(L)
    print(lst1)

def mergesort_by_character(L):
    # Base case: A list of 0 or 1 elements is already sorted
    if len(L) <= 1:
        return L
    
    # Step 1: Split the list into two halves
    mid = len(L) // 2
    left_half = L[:mid]
    right_half = L[mid:]
    
    # Step 2: Recursively sort both halves
    left_sorted = mergesort_by_character(left_half)
    right_sorted = mergesort_by_character(right_half)
    
    return merge1(left_sorted, right_sorted)
    
def merge1(left, right):
    i = j = 0
    sorted_list = []
    
    while i < len(left) and j < len(right):
        if left[i][0] < right[j][0]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1
    while i < len(left):
        sorted_list.append(left[i])
        i += 1
    
    while j < len(right):
        sorted_list.append(right[j])
        j += 1
    
    return sorted_list
```

```python
lst = list(map(str, input().split(', ')))
combinationSort(lst)
```
